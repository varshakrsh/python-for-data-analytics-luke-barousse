```python
from datasets import load_dataset
dataset= load_dataset('lukebarousse/data_jobs')
```

```python
df= dataset['train'].to_pandas()
df
```

```python
df.info()
```

```python
import pandas as pd

df.job_posted_date= pd.to_datetime(df.job_posted_date)
df.info()
```

```python
df['job_posted_month']= df.job_posted_date.dt.month
df
```

```python
df.sort_values('job_posted_date', inplace=True)
df
```

```python
df['job_country'].unique()
```

```python
sg_jobs=df[df['job_country']=='Singapore']
sg_jobs
```

```python
ba_sg=sg_jobs[sg_jobs['job_title_short']=='Business Analyst']
ba_sg
```

```python
ba_sg[ba_sg['salary_year_avg'].notna()]
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df['job_month']=df['job_posted_date'].dt.month
df
```

```python
y=df['job_month'].value_counts()
y= y.sort_index()
plt.plot(y.index, y)
```

```python
job_postings=df.job_title_short.value_counts()
plt.bar(job_postings.index, job_postings)
plt.title('job postings')
plt.ylabel('no. of postings')
plt.xlabel('job title')
plt.xticks(rotation=90)
```

```python
job_postings.plot(kind='bar')
plt.title('job postings')
plt.xlabel('job titles')
plt.ylabel('no of postings')
plt.xticks(rotation=45, ha='right')
```

```python
jobs_salaries= df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
jobs_salaries.info()
```

```python
plt.barh(jobs_salaries.index, jobs_salaries)
plt.title('median salaries by jobs')
plt.ylabel('job title')
plt.xlabel('median salary')
```
