```python
#loading the dataset
from datasets import load_dataset
dataset= load_dataset('lukebarousse/data_jobs')
```

```python
#converting the dataset into a dataframe
df= dataset['train'].to_pandas()
df
```

```python
#obtaining information about the dataframe using .info()
df.info()
```

```python
#importing pandas and fomatting the job_posted_date column
import pandas as pd

df.job_posted_date= pd.to_datetime(df.job_posted_date)
df.info()
```

```python
#obtaining months from the dates
df['job_posted_month']= df.job_posted_date.dt.month
df
```

```python
#sorting the dataframe in the order of posted dates
df.sort_values('job_posted_date', inplace=True)
df
```

```python
#finding all the countries listed in the dataframe
df['job_country'].unique()
```

```python
#Filtering jobs in Singapore
sg_jobs=df[df['job_country']=='Singapore']
sg_jobs
```

```python
#Filtering business analyst jobs in Singapore
ba_sg=sg_jobs[sg_jobs['job_title_short']=='Business Analyst']
ba_sg
```

```python
#Filtering out non-NA salary values
ba_sg[ba_sg['salary_year_avg'].notna()]
```

```python
#importing matplotlib and numpy
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creating a month column
df['job_month']=df['job_posted_date'].dt.month
df
```

```python
#plotting the no.of jobs posted through each month in a line graph
y=df['job_month'].value_counts()
y= y.sort_index()
plt.plot(y.index, y)
plt.show()
```

```python
#plotting a barplot that shows the no.of listings for each job
job_postings=df.job_title_short.value_counts()
plt.bar(job_postings.index, job_postings)
plt.title('job postings')
plt.ylabel('no. of postings')
plt.xlabel('job title')
plt.xticks(rotation=90)
plt.show()
```

```python
#plotting the no.of listings for each job using .plot()
job_postings.plot(kind='bar')
plt.title('job postings')
plt.xlabel('job titles')
plt.ylabel('no of postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
#grouping by job title and median salary, sorting values based on median salary
jobs_salaries= df.groupby('job_title_short')['salary_year_avg'].median().sort_values()
jobs_salaries.info()
```

```python
#plotting a bar plot that shows the median salary for each job
plt.barh(jobs_salaries.index, jobs_salaries)
plt.title('median salaries by jobs')
plt.ylabel('job title')
plt.xlabel('median salary')
plt.show()
```
