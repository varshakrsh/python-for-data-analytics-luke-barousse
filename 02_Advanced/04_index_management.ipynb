```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df

```

```python
#assigning a name for the index

df.index.name='index_name'
df
```

```python
#Resetting Index using .reset_index()
df_new= df[df['job_country'] == 'Singapore']
df_new.reset_index(inplace=True)
df_new
```

```python
#setting a particular column as the index using .set_index()
df_new.set_index('index_name', inplace= True)
df_new
```

```python
#Sorting a dataframe based on the indices using .sort_index()
df_new.sort_index(ascending=False, inplace=True)
df_new
```
