```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
#returning a random sample of entries from the dataframe using .sample()
df.sample(5)
```

```python
#To obtain the same set of entries everytime .sample() is ran
df.sample( 5, random_state=42)
```

```python
#Creating a copy of the original dataframe without modifying the original dataframe
copy_df= df.copy()
id_df= id(df)
id_copy= id(copy_df)
print(id_df)
print(id_copy)
```

As the IDs of both dataframes are different, they are 2 different objects. Modifying the copy won't affect the original dataframe.
