```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
#Creating a month column in the dataframe
df['job_posted_month']=df['job_posted_date'].dt.strftime('%b')
df
```

```python
#splitting the dataframe based on months
months=df['job_posted_month'].unique()
month_dict={month:df[df['job_posted_month']==month] for month in months}

```

```python
#creating a dataframe for the first quarter using .concat()
new_df=pd.concat([month_dict['Jan'], month_dict['Feb'], month_dict['Mar']], ignore_index=True)
new_df
```

```python
#Plotting a graph to show the no. of job postings during the first quarter
new_df['job_posted_month'].value_counts().plot(kind='bar')
plt.xlabel('Fisrt quarter')
plt.ylabel('no. of postings')
plt.title('job postings during the first quarter')
plt.show()
```
