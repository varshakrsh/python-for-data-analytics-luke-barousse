```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
#filtering jobs in Singapore
sg_jobs=df[df['job_country']=='Singapore']
sg_jobs
```

```python
#finding the top 3 most listed jobs
top_jobs= sg_jobs.job_title_short.value_counts().head(3).index
top_jobs
```

```python
#creating a month column in the dataframe
sg_jobs['job_month']=sg_jobs['job_posted_date'].dt.strftime('%B')
```

```python
#creating a pivot table that shows the no.of job postings by month
pivot_df= sg_jobs.pivot_table(index='job_month', columns='job_title_short', aggfunc='size')
pivot_df
```

```python
#Resetting the index of the pivot table 
pivot_df.reset_index(inplace=True)
```

```python
#Creating a column for month no and sorting them in order
pivot_df['month_no']=pd.to_datetime(pivot_df['job_month'], format='%B').dt.month
pivot_df.sort_values('month_no', inplace=True)
```

```python
#dropping the month_no column
pivot_df.drop(columns='month_no',inplace=True)
```

```python
#setting month as the index
pivot_df.set_index('job_month',inplace=True)
```

```python
pivot_df
```

```python
#plotting a line graph
pivot_df=pivot_df[top_jobs]
pivot_df.plot(kind='line')
plt.xlabel('month')
plt.ylabel('no.of postings')
plt.title('monthly postings for the top 3 jobs in Singapore')
plt.show()
```
