```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
#filtering out data analyst jobs in United states and dropping NA values in salary column
df_new=df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')]
df_new=df_new.dropna(subset='salary_year_avg')
```

```python
#creating a box plot that shows the distribution of yearly salaries in the US for data analyst jobs
df_new['salary_year_avg'].plot(kind='box',vert=False)
plt.show()
```

```python
# Filtering data analyst, data engineer, and data scientist jobs in the US
jobs=['Data Analyst', 'Data Engineer', 'Data Scientist']
df_jobs=df[(df['job_title_short'].isin(jobs)) & (df['job_country']=='United States')]
```

```python
#creating a box plot that shows the distribution of salaries for the 3 jobs in the US
#dropping NA values in salary column
df_jobs=df_jobs.dropna(subset='salary_year_avg')
jobs_list=[df_jobs[df_jobs['job_title_short']==job]['salary_year_avg'] for job in jobs]
plt.boxplot(jobs_list, labels=jobs, vert=False)
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.title('distribution of salaries for data jobs in the US')
plt.xlim(0,600000)
plt.xlabel('salaries')
plt.ylabel('job titles')
plt.show()
```
