```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
#cleaning the job_skills column by converting them from strings into lists
import ast
df['job_skills']=df['job_skills'].apply(lambda skills: ast.literal_eval(skills) if pd.notna(skills) else skills )
```

```python
#Filtering out data analyst jobs in the US
df_new=df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')]
```

```python
#exploding skills column
df_new=df_new.explode('job_skills')
```

```python
#creating a new dataframe that groups skill count and median yearly salary
df_skills=df_new.groupby('job_skills')['salary_year_avg'].agg(['count','median'])
```

```python
#creating 2 dataframes - one of them for highest paid skills and the other for skills in demand
df_salary=df_skills.sort_values(by='median', ascending=False).head(10)
df_demand=df_skills.sort_values(by='count',ascending=False).head(10).sort_values(by='median', ascending=False)
```

```python
#plotting a bar graph that compares these 2 dataframes
fig,ax=plt.subplots(2,1)
df_salary.plot(kind='barh', y='median', ax=ax[0], legend=False)
ax[0].invert_yaxis()
ax[0].set_xlabel('')
ax[0].set_ylabel('')
ax[0].set_title('top 10 most paid skills for data analysts in the US')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
df_demand.plot(kind='barh', y='median',ax=ax[1], legend=False)
ax[1].invert_yaxis()
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('median salary')
ax[1].set_title('top 10 in demand skills for data analysts in the US')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.tight_layout()
```

```python
# customising this plot using seaborn
import seaborn as sns
fig,ax=plt.subplots(2,1)
sns.set_theme(style='ticks')
sns.barplot(df_salary, y=df_salary.index, x='median', ax=ax[0], hue='median', palette='dark:b_r')
ax[0].legend().remove()
ax[0].set_xlabel('')
ax[0].set_ylabel('')
ax[0].set_title('top 10 most paid skills for data analysts in the US')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
sns.barplot(df_demand, x='median', y=df_demand.index, ax=ax[1], hue='median', palette='dark:b_r')
ax[1].legend().remove()
ax[1].set_xlim(ax[0].get_xlim())
ax[1].set_ylabel('')
ax[1].set_xlabel('median salary')
ax[1].set_title('top 10 in demand skills for data analysts in the US')
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.tight_layout()
```

```python
# customizing histograms using Seaborn
#plotting a histogram that shows the distribution of yearly salaries for data analysts in the US
df_new['salary_year_avg'].plot(kind='hist',bins=30,edgecolor='black')
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.xlabel('yearly salaries')
plt.xlim(0,250000)
plt.ylabel('frequency')
plt.title('distribution of yearly salaries for data analysts in the US')
plt.show()
```

```python
#customising the histogram above to get a distribution curve using .displot()
sns.displot(df_new['salary_year_avg'], kind='kde', fill=True)
sns.set_theme(style='ticks')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.xlabel('yearly salaries')
plt.xlim(0,250000)
plt.ylabel('frequency')
plt.title('distribution of yearly salaries for data analysts in the US')
plt.show()
```

```python
#customizing box plots using seaborn
# Filtering data analyst, data engineer, and data scientist jobs in the US
jobs=['Data Analyst', 'Data Engineer', 'Data Scientist']
df_jobs=df[(df['job_title_short'].isin(jobs)) & (df['job_country']=='United States')]
```

```python
#creating a box plot that shows the distribution of salaries for the 3 jobs in the US
#dropping NA values in salary column
df_jobs=df_jobs.dropna(subset='salary_year_avg')
jobs_list=[df_jobs[df_jobs['job_title_short']==job]['salary_year_avg'] for job in jobs]
plt.boxplot(jobs_list, labels=jobs, vert=False)
ax=plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.title('distribution of salaries for data jobs in the US')
plt.xlim(0,600000)
plt.xlabel('salaries')
plt.ylabel('job titles')
plt.show()
```

```python
# customizing the box plot created above
sns.boxplot(df_jobs,x='salary_year_avg', y='job_title_short')
sns.set_theme(style='ticks')
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f'${int(x/1000)}K'))
plt.title('distribution of salaries for data jobs in the US')
plt.xlim(0,600000)
plt.xlabel('salaries')
plt.ylabel('job titles')
plt.show()
```
