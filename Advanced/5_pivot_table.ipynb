```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')

#converting the loaded dataset into a pandas dataframe
df=dataset['train'].to_pandas()

#formatting datetime
df['job_posted_date']=pd.to_datetime(df['job_posted_date'])
df
```

```python
# finding the top 5 countries with the highest number of postings
top_countries=df.job_country.value_counts().head(5).index
top_countries
```

```python
#finding the three most listed jobs
top_jobs= df.job_title_short.value_counts().head(3).index
top_jobs
```

```python
#creating a pivot table that combines median salary by top 5 countries and top 3 jobs using .pivot_table()

df_pivot= df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short', aggfunc='median')
df_pivot= df_pivot.loc[top_countries]
df_pivot= df_pivot[top_jobs]
df_pivot

```

```python
#plotting a bar chart for the pivot table
df_pivot.plot(kind= 'bar')
plt.ylabel('Median salary in USD')
plt.xlabel('Top 5 countries')
plt.title(' Median salary by top 5 countries and top 3 jobs')
plt.show()
```
